## Preprocessing

Création d'un objet cook contenant : 

1. **un preprocessing :** 

    1. Un Regex permettant de nettoyer le jeu de données de toutes les lignes comprenant des caractères spéciaux ( chiffres, . ou autre...) 
    2. A partir de la bibliotheque nltk, application de l'outil pos tag permettant l'étiquetage **morpho syntaxique** puis conservation des : noms, adverbes et verbes. 
    3. Application d'un multilabelizer permettant de créer des features basées sur les listes d'ingrédients néttoyées.
    
2. **une modélisation :** 
    1. Des modèles à tester (SVC, Logistic Regression...)
    2. Optimisation des parametres (random search...)
    3. Ensemble learning (Stacking, Boosting...)
    
    
Sans application des Regex et du pos tag,le preprocessing créait 6700 features, 6400 avec les Regex, puis 4000 avec pos tag

## Test de la performance d'un modèle SVC 

### Test model SVC via méthode train test split

```python
simple_cook = Cook('train.json')
simple_cook.test_model(LinearSVC()) 
```

**Interpretation du résultat :** 
Nous avons une accuracy de 76% (comparaison de nos prédictions vs. données test). Cette métrique est celle utilisée dans le challenge Kaggle. 

Cela signifie que 76% des valeures prédites sont correctes. 

### Test model SVC via méthode cross validation

```python
cross_cook = Cook('train.json')
cross_cook.cross_model(LinearSVC()) 
```

Nous utiliserons un TrainTest Split figé sur un random et non pas une cross validation pour évaluer nos résultats compte tenu de la durée des tests.

## Tests des modèles simples

```python
simple_models = (
    LinearSVC(),
    SGDClassifier(),
    KNeighborsClassifier(),
    LogisticRegression()
)

for simple in simple_models:
    simple_cook = Cook('train.json')
    simple_cook.test_model(simple) 
```

**Interpretation du résultat :**

* Pour le modèle SGDClassifier, nous avons une accuracy de 77,6%. Le résultat est amélioré de 1%.
* Pour le modèle KNeighborsClassifier, la performance du modele est fortement impactée par le nombre important de features
* Pour le modèle LogisticRegression, le résultat équivalent à SGD Classifier

## Random search on SGD Classifier 
L'objectif est d'optimiser les parametres du meilleur modele (SGD Classifier)

```python
randomize_SGD_cook = Cook('train.json')
randomize_SGD_cook.randomize_training()
```

## Test on Best Estimator's SGDClassifier

```python
best_SGD_cook = Cook('train.json')
best_SGD_cook.test_model(SGDClassifier(penalty='l2', n_jobs=-1, alpha=0.0001))
```

## Test des ensembles Learning pour améliorer le modèle

```python
ensemble_models = (
    GradientBoostingClassifier(n_estimators=10),
    AdaBoostClassifier(n_estimators=50, learning_rate=1.0)    
)

for ensemble in ensemble_models:
    ensemble_cook = Cook('train.json')
    ensemble_cook.test_model(ensemble) 
```

Les résultats sont inférieurs aux modèles simples

## Test on Stacking

```python
stacking_cook = Cook('train.json')
stacking_cook.stacking_training(LinearSVC(), SGDClassifier(penalty='l2', n_jobs=-1, alpha=0.0001))
```

Les résultats sont plus ou moins équivalents avec les modèles simples avec néanmoins un très fort temps d'éxecution
